```python

import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

# Sample data (replace with your dataset)
X_train = np.random.rand(100, 5)  # Example: 100 samples, 5 features
y_train = np.random.rand(100, 1)  # Example: 100 samples, 1 output (salary)

# Build RNN model
model = Sequential([
    SimpleRNN(64, input_shape=(X_train.shape[1], 1)),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')

# Train model
model.fit(X_train, y_train, epochs=10)

# Generate a sample input for prediction
X_sample = np.random.rand(1, 5)

# Predict salary using the trained model
predicted_salary = model.predict(X_sample)
print("Predicted Salary:", predicted_salary[0][0])
```
